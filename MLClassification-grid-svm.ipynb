```python
import pandas  as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
num = dataset.select_dtypes(include= ['number']).columns.to_list()
cat = dataset.select_dtypes(include= ['category','object']).columns.to_list()
#print(num)
print(cat)
```

```python
dataset=pd.get_dummies(dataset,drop_first="True")
```

```python
dataset.shape
```

```python
input=dataset.drop(columns=["classification_yes"])
```

```python
input
```

```python
dataset[num].isna().sum()
```

```python
output=dataset["classification_yes"]
```

```python
print(output)
```

```python
from sklearn.model_selection import train_test_split
#from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(input,output,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'kernel':['linear','rbf','poly','sigmoid'],'gamma':['auto','scale'],'C':[10,100,1000,2000,3000]} 



grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
# fitting the model for grid search 
grid.fit(x_train, y_train) 
 



#from sklearn.model_selection import GridSearchCV

#param_grid = {'kernel':['linear','rbf','poly','sigmoid'],
           #  'gamma':['auto','scale'],
            # 'C':[10,100,1000,2000,3000]} 



#grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
# fitting the model for grid search 
#grid.fit(y_train, y_train) 
 
```

```python
re=grid.cv_results_
#print(re)
grid_predictions = grid.predict(x_test) 
   

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)



# print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)







#from sklearn.metrics import confusion_matrix

#cm=confusion_matrix(y_test,y_pred)
```
